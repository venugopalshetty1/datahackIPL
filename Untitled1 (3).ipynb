```python
batsman_details=ball_by_ball[['match_id','inning','batting_team','batsman','batsman_runs']]
batsman_details=batsman_details.merge(match_data[['match_id','season']],on="match_id")
```

```python
batsman_runs=batsman_details.groupby(['match_id','season','inning','batting_team','batsman'])['batsman_runs'].agg({'balls_played':'count','batsman_runs':'sum'}).reset_index()
```

```python
#### Calculate the strike rate for every season than every match


batsman_runs_season=batsman_runs.groupby(['batsman'])['balls_played','batsman_runs'].agg({'balls_played':'sum','batsman_runs':'sum'}).reset_index()
```

```python
batsman_runs_season['strick_rate']=np.round(batsman_runs_season['batsman_runs']/batsman_runs_season['balls_played']*100,2)
```

```python
#### Ploting a scatter plot for the number of runs scored vs strike rate of the batsmen
from plotly.graph_objs import Scatter

trace0 = Scatter(x=batsman_runs_season['batsman_runs'],
                 y=batsman_runs_season['strick_rate'],
                 mode='markers',
                 marker=dict(size=9, color='rgba(152, 0, 0, .8)', line=dict(width=1)),
                 text=batsman_runs_season['batsman']
                 )
layout=Layout(title='Batsman Strick rate VS total run Scored',
              xaxis=dict(title="Total runs scored"),
              yaxis=dict(title="Over all Strick Rate"))


### Offline Scatter plot
### Thiz is JSON format

plotly.offline.iplot({
    "data":[trace0],
    "layout":layout
})

```

```python
#### Over all bowler economy calculation..
#### coverting the overs to the actual decimals 

#### Formula: over_all_run_given/no_of_overs_bowled
#### preparing the data


bowler_details=ball_by_ball[['match_id','bowling_team','over','bowler','batsman_runs','extra_runs','player_dismissed','player_dismissed_id','dismissal_kind'
]]
bowler_details.head()

```

```python
bowler_overs=pa.DataFrame(bowler_details.groupby(['match_id','bowling_team','over','bowler'])['over','batsman_runs'].agg({"over":"count","batsman_runs":"sum"}))
bowler_overs.columns=['no_of_balls','runs_per_over']
bowler_overs=bowler_overs.reset_index()
bowler_overs['extras']=0
```

```python
#### calculating the extras bowled per over

for i in range(len(bowler_overs)):
    if bowler_overs.iloc[i,4]>6:
        bowler_overs.iloc[i,6]=bowler_overs.iloc[i,4]-6
        bowler_overs.iloc[i,4]=6
```

```python
bowler_overs=bowler_overs.groupby(['bowler'])['no_of_balls','runs_per_over','extras'].agg({'no_of_balls':"sum","runs_per_over":"sum","extras":"sum"}).reset_index()
bowler_overs.columns=['bowler','overs','run_given','total_extras']
bowler_overs['overs']=np.round(bowler_overs['overs']/6,1)
```

```python
#### Calculation of economy for the bowlers 
#### formula: total_runs_given/overs_bowled


bowler_overs['economy']=np.round(bowler_overs['run_given']/bowler_overs['overs'],3)
```

```python
bowler_overs.head()
```

```python
####Wickets taken by the bowler if the dismissal kind is caught,bowled,lbw,hit wicket, stumped, caught and bowled
dismissal=['caught','bowled','lbw','hit wicket','stumped','caught and bowled']
wickets=bowler_details[bowler_details['dismissal_kind'].isin(dismissal)][['bowler','dismissal_kind']]
wickets=(wickets['bowler']).value_counts()
wickets=wickets.reset_index()
wickets.columns=['bowler','wickets']
print(wickets.head())
bowler_overs=bowler_overs.merge(wickets, on="bowler")
bowler_overs.head()
```

```python
bowler_overs[(bowler_overs['economy']>4.0) & (bowler_overs['economy']<7.0)]
```

```python
#### plotting the economy of the bowlers. So that during the acution.
#### Segregating the different colors' for very expensive, expensive, economical, very economical
#### Conditions:   very economical <7.0
####               economical >=7.0 and <=10.0
####               expensive >10.0 and <=12.0
####               very expensive >12.0 


Very_Economical = Scatter(x=bowler_overs[bowler_overs['economy']<7.0]['wickets'],
                 y=bowler_overs[bowler_overs['economy']<7.0]['economy'],
                 mode="markers",
                 marker=dict(size=9, color='rgba(152, 0, 0, .8)', line=dict(width=1)),
                 text=bowler_overs[bowler_overs['economy']<7.0]['bowler'],
                 name="Very_economical")

Economical = Scatter(x=bowler_overs[(bowler_overs['economy']>=7.0) & (bowler_overs['economy']<=10.0)]['wickets'],
                 y=bowler_overs[(bowler_overs['economy']>=7.0) & (bowler_overs['economy']<=10.0)]['economy'],
                 mode="markers",
                 marker=dict(size=9, color='rgba(100, 255, 0, .8)', line=dict(width=1)),
                 text=bowler_overs[(bowler_overs['economy']>=7.0) & (bowler_overs['economy']<=10.0)]['bowler'],
                 name="Economical")


Expensive = Scatter(x=bowler_overs[(bowler_overs['economy']>10.0) & (bowler_overs['economy']<=12.0)]['wickets'],
                 y=bowler_overs[(bowler_overs['economy']>10.0) & (bowler_overs['economy']<=12.0)]['economy'],
                 mode="markers",
                 marker=dict(size=9, color='rgba(100, 100, 0, .8)', line=dict(width=1)),
                 text=bowler_overs[(bowler_overs['economy']>10.0) & (bowler_overs['economy']<=12.0)]['bowler'],
                 name="Expensive")

Very_Expensive = Scatter(x=bowler_overs[bowler_overs['economy']>12.0]['wickets'],
                 y=bowler_overs[bowler_overs['economy']>12.0]['economy'],
                 mode="markers",
                 marker=dict(size=9, color='rgba(255, 100, 0, .8)', line=dict(width=1)),
                 text=bowler_overs[bowler_overs['economy']>12.0]['bowler'],
                 name="Very_expensive")

layout = Layout(title="Bowler Economy VS total wickets taken",
                xaxis=dict(title="Total wickets taken in series"),
                yaxis=dict(title="Over all Economy"))

plotly.offline.iplot({
    "data"  :[Very_Economical,Economical,Expensive,Very_Expensive],
    "layout":layout
})
```

```python
#### Here we compared How many matches the team played overall and 
#### how many times they own the toss in terms of percentage it ploted.

Toss_winning= Bar(x=matchs_played['team'],
            y=matchs_played['toss_win_matches'],
            marker=dict(color="rgb(158,202,225)"),
            name='Toss_win',
            text=matchs_played['prcnt_toss_winning'])
Matches_played = Bar(x=matchs_played['team'],
             y=matchs_played['matches'],
             marker=dict(color='rgb(100,200,120)'),
             name='Matches_played')

layout = Layout(title="Matches Played VS Toss Winning",
                xaxis=dict(title="Teams"),
                yaxis=dict(title="Number of Matches"))

plotly.offline.iplot({
    "data":[Matches_played,Toss_winning],
    "layout":layout
})
```
