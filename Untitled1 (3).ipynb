```python
#### Over all bowler economy calculation..
#### coverting the overs to the actual decimals 

#### Formula: over_all_run_given/no_of_overs_bowled
#### preparing the data


bowler_details=ball_by_ball[['match_id','bowling_team','over','bowler','batsman_runs','extra_runs','player_dismissed','player_dismissed_id','dismissal_kind'
]]
bowler_details.head()

```

```python
bowler_overs=pa.DataFrame(bowler_details.groupby(['match_id','bowling_team','over','bowler'])['over','batsman_runs'].agg({"over":"count","batsman_runs":"sum"}))
bowler_overs.columns=['no_of_balls','runs_per_over']
bowler_overs=bowler_overs.reset_index()
bowler_overs['extras']=0
```

```python
#### calculating the extras bowled per over

for i in range(len(bowler_overs)):
    if bowler_overs.iloc[i,4]>6:
        bowler_overs.iloc[i,6]=bowler_overs.iloc[i,4]-6
        bowler_overs.iloc[i,4]=6
```

```python
bowler_overs=bowler_overs.groupby(['bowler'])['no_of_balls','runs_per_over','extras'].agg({'no_of_balls':"sum","runs_per_over":"sum","extras":"sum"}).reset_index()
bowler_overs.columns=['bowler','overs','run_given','total_extras']
bowler_overs['overs']=np.round(bowler_overs['overs']/6,1)
```

```python
#### Calculation of economy for the bowlers 
#### formula: total_runs_given/overs_bowled


bowler_overs['economy']=np.round(bowler_overs['run_given']/bowler_overs['overs'],3)
```

```python
bowler_overs.head()
```

```python
####Wickets taken by the bowler if the dismissal kind is caught,bowled,lbw,hit wicket, stumped, caught and bowled
dismissal=['caught','bowled','lbw','hit wicket','stumped','caught and bowled']
wickets=bowler_details[bowler_details['dismissal_kind'].isin(dismissal)][['bowler','dismissal_kind']]
wickets=(wickets['bowler']).value_counts()
wickets=wickets.reset_index()
wickets.columns=['bowler','wickets']
print(wickets.head())
bowler_overs=bowler_overs.merge(wickets, on="bowler")
bowler_overs.head()
```

```python
#### above we seen the observed values 
#### now calculating the expected values.
#### here chi square is used to calculate the association between the two variables ---->



"""
observed values:

         0       1          row total
bat     149     124           273
  
field   162     201           463

                              
column  311     325           636     over all total
total


"""


"""

Expected values:

i=row
j=column

formula:   E(i,j)  =   (row_total(i)*column_total(j)) / over_all_total
 


           0         1  

bat      133.49     139.50
field    177.50     185.49




chi_square_statistic =  sum(((observed - expected)^2) / expected)

after getting chi square
degree of freedom is calcualted 

(r-1)*(c-1)  -----------> r=number of rows , c = number of columns in cross tabulation.


important note: chi square distrubution is not a symmetric distribution, it is little skewed to right.

degress of freedom = 1


"""

expected_values=[133.49,139.50,226.40,236.59]
observed_values=[[149,124],[162,201]]

from scipy.stats import chi2_contingency


```

```python
### what is the trend of the sixes based on the years.



sixes_trend=ball_by_ball[(ball_by_ball['inning']==1) | (ball_by_ball['inning']==2)][['match_id','batting_team','inning','batsman','batsman_runs','extra_runs']]
```

```python
sixes_trend['season']=[row.split("_")[0] for row in sixes_trend['match_id']]
```

```python
six=sixes_trend[sixes_trend['batsman_runs']==6].groupby(['season'], as_index=False)['batsman_runs'].count()
four=sixes_trend[sixes_trend['batsman_runs']==4].groupby(['season'], as_index=False)['batsman_runs'].count()
ones=sixes_trend[sixes_trend['batsman_runs']==1].groupby(['season'], as_index=False)['batsman_runs'].count()
twos=sixes_trend[sixes_trend['batsman_runs']==2].groupby(['season'], as_index=False)['batsman_runs'].count()
threes=sixes_trend[sixes_trend['batsman_runs']==3].groupby(['season'], as_index=False)['batsman_runs'].count()
```

```python
trace0 = Scatter(x=six['season'],
              y=six['batsman_runs'],
              mode="lines+markers",
              name="Sixes")

trace1 = Scatter(x=four['season'],
              y=four['batsman_runs'],
              mode="lines+markers",
              name="Fours")
trace2 = Scatter(x=ones['season'],
              y=ones['batsman_runs'],
              mode="lines+markers",
              name="ones")
trace3 = Scatter(x=twos['season'],
              y=twos['batsman_runs'],
              mode="lines+markers",
              name="Twos")
trace4 = Scatter(x=threes['season'],
              y=threes['batsman_runs'],
              mode="lines+markers",
              name="Threes")



layout=Layout(title="Trend line for ones,twos, threes, fours, sixes",
              xaxis=dict(title="seasons"),
              yaxis=dict(title="Trends"))


plotly.offline.iplot({
    "data":[trace0,trace1,trace2,trace3,trace4],
    "layout":layout
})

```
