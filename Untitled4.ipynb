```python
####Visualizations
#### Importing dependencies 
import numpy as np
import pandas as pa
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
ball_by_ball = pa.read_csv("./train/train/ball_by_ball_data.csv")
match_data = pa.read_csv("./train/train/match_data.csv")
player_roster = pa.read_csv("./train/train/player_rosters.csv")
key_teams = pa.read_csv("./train/train/key_teams.csv")
matchs_2018=pa.read_csv("./test/test/matches_2018.csv")
player_prediction_2018=pa.read_csv("./test/test/player_predictions.csv")
```

```python
filter_data=ball_by_ball[['match_id','inning','batting_team','bowling_team','over','ball','bowler','bowler_id','batsman_runs','extra_runs','dismissal_kind']]
filter_data=filter_data.merge(match_data[['match_id','venue_id','season']],on='match_id')
```

```python
players_2018=player_roster[player_roster['Season']==2018][['player_id','Player','Team']].reset_index()
del players_2018['index']
print(players_2018.head(10))
```

```python
#### Replacing Team DC with SRH.

team_change={'DC':'SRH'}


filter_data=filter_data.replace({'batting_team':team_change,'bowling_team':team_change})
```

```python
#### I want only the bowlers in 2018.

filter_data_2018=filter_data[filter_data['bowler_id'].isin(list(players_2018['player_id']))]
```

```python
filter_data_2018['runs']=filter_data_2018['batsman_runs']+filter_data_2018['extra_runs']
```

```python
filter_data_2018
```

```python
#### seperating first and second innings.

filter_data_1st_inn = filter_data_2018[filter_data_2018['inning']==1]
filter_data_2nd_inn = filter_data_2018[filter_data_2018['inning']==2]
```

############################################################################################################################

# First innings wickets

```python
balls_runs_1st_inn=filter_data_1st_inn.groupby(['match_id','season','venue_id','batting_team','bowling_team','bowler_id'], as_index=False)['runs'].agg({'runs':'sum','ball_bowled':'count'})
```

```python
balls_runs_1st_inn[balls_runs_1st_inn['bowler_id']==3]['match_id'].unique()
```

```python
dismissal_kind=['caught', 'bowled', 'lbw', 'caught and bowled','stumped', 'hit wicket']
```

```python
wickets_1st_inn=filter_data_1st_inn[filter_data_1st_inn['dismissal_kind'].isin(dismissal_kind)]
```

```python
wickets_1st_inn['dismissal_kind']=1
```

```python
wickets_1st_inn=wickets_1st_inn.groupby(['match_id','batting_team','bowling_team','bowler_id'],as_index=False)['dismissal_kind'].sum()
```

```python
wickets_1st_inn
```

```python
final_1st_inn_data=balls_runs_1st_inn.merge(wickets_1st_inn,on=['match_id','bowler_id','batting_team','bowling_team'],how='left')
```

```python
final_1st_inn_data[final_1st_inn_data['bowler_id']==3]['match_id'].unique()
```

```python
#### Adding the seasonal match_id.

final_1st_inn_data['seasonal_match_id']=[int(row.split('_')[1]) for row in final_1st_inn_data['match_id']]
final_1st_inn_data['season']=(final_1st_inn_data['season']).apply(int)
```

```python
final_1st_inn_data['dismissal_kind']=final_1st_inn_data['dismissal_kind'].fillna(0)
```

```python
####Sorting ascending of matches for each player.

final_1st_inn_data=final_1st_inn_data.sort_values(['bowler_id','season','seasonal_match_id'],ascending=[True,True,True])
```

```python
final_1st_inn_data[final_1st_inn_data['bowler_id']==3]['match_id']
```

```python
#### Summing up to ball bowled (cum_sum).

final_1st_inn_data['total_balls_bowled']=final_1st_inn_data.groupby(['bowler_id'])['ball_bowled'].cumsum()
```

```python
#### Generating the total_matches_played.

final_1st_inn_data['total_matches_played']=(final_1st_inn_data.groupby(['bowler_id'])['total_balls_bowled'].rank()).apply(int)
```

```python
#### Generating the total_runs_given.

final_1st_inn_data['total_runs_given']=final_1st_inn_data.groupby(['bowler_id'])['runs'].cumsum()
```

```python
#### Generating the cummulative sum of wickets for the each match of each bowler.

final_1st_inn_data['total_wickets_taken']=final_1st_inn_data.groupby(['bowler_id'])['dismissal_kind'].cumsum()
```

```python
### Averge of a bowler.
### formula:  Run_given/wickets_taken.
### lower the bowling average better the bowler is performing.


final_1st_inn_data['bowling_average']=np.round(final_1st_inn_data['total_runs_given']/final_1st_inn_data['total_wickets_taken'],2)
```

```python
### Strick rate of bowler.
### formula: overall_balls_bowled/wickets_taken.
### Lower the strike rate bowler is preforming well. Means he is taking the wickets so quickly.



final_1st_inn_data['bowling_strike_rate']=np.round(final_1st_inn_data['total_balls_bowled']/final_1st_inn_data['total_wickets_taken'],2)
```

```python
#### calculate number of overs played by the bowler.
overs_details=filter_data_1st_inn[['match_id','season','bowler_id','over','ball']].groupby(['match_id','season','bowler_id','over'],as_index=False).count()
```

```python
for i in range(len(overs_details)):
    if overs_details.iloc[i,4]>6:
        overs_details.iloc[i,4]=6

del overs_details['over']   
```

```python
overs_details=overs_details.groupby(['match_id','season','bowler_id'],as_index=False)['ball'].sum()
overs_details['ball']=np.floor(overs_details['ball']/6)
```

```python
overs_details.columns=['match_id','season','bowler_id','overs_played']
```

```python
overs_details
```

```python
final_1st_inn_data=final_1st_inn_data.merge(overs_details,on=['match_id','season','bowler_id'])
```

```python
### Economy rate calculation


###1.sort again the order.

final_1st_inn_data=final_1st_inn_data.sort_values(['bowler_id','season','seasonal_match_id'])


```

```python
final_1st_inn_data
```

```python
###2. calculate the feature cummulative sum of overs for each player.


final_1st_inn_data['cum_sum_overs']=final_1st_inn_data.groupby(['bowler_id'])['overs_played'].cumsum()
```

```python
### Calculation of economy rate of the player for each match cummulatively.
### Formula: total_run_given/total_overs_bowled.


final_1st_inn_data['economy_player']=np.round(final_1st_inn_data['total_runs_given']/final_1st_inn_data['cum_sum_overs'],2)

```

```python
##### Replace the inf values with 0 since there is no wickets taken so zero bowling average and strike rate

for i in (final_1st_inn_data[final_1st_inn_data['bowling_average'].isin([np.inf])].index):
    final_1st_inn_data.iloc[i,14]=0
    final_1st_inn_data.iloc[i,15]=0
```

```python
#### remove some of unwanted variables.

del final_1st_inn_data['ball_bowled']

del final_1st_inn_data['runs']
```

```python
validation_2018_1st_inn=final_1st_inn_data[final_1st_inn_data['bowler_id'].isin(players_2018['player_id'])]
```

```python
validation_2018_1st_inn=validation_2018_1st_inn.sort_values(['bowler_id','season','seasonal_match_id'],ascending=[True,False,False]).drop_duplicates(subset='bowler_id',keep='first')
```

```python
#### shifting total_balls_bowled.
final_1st_inn_data['total_balls_bowled']=final_1st_inn_data.groupby(['bowler_id'])['total_balls_bowled'].shift(1)
```

```python
#### shifting total_matches_played.
final_1st_inn_data['total_matches_played']=final_1st_inn_data.groupby(['bowler_id'])['total_matches_played'].shift(1)
```

```python
#### shifting total_runs_given.
final_1st_inn_data['total_runs_given']=final_1st_inn_data.groupby(['bowler_id'])['total_runs_given'].shift(1)
```

```python
#### shifting bowling_average.
final_1st_inn_data['bowling_average']=final_1st_inn_data.groupby(['bowler_id'])['bowling_average'].shift(1)
```

```python
#### shifting bowling_strike_rate.
final_1st_inn_data['bowling_strike_rate']=final_1st_inn_data.groupby(['bowler_id'])['bowling_strike_rate'].shift(1)
```

```python
#### shifting cum_sum_overs.
final_1st_inn_data['cum_sum_overs']=final_1st_inn_data.groupby(['bowler_id'])['cum_sum_overs'].shift(1)
```

```python
#### shifting economy_player.
final_1st_inn_data['economy_player']=final_1st_inn_data.groupby(['bowler_id'])['economy_player'].shift(1)
```

```python
##### leaving the very first match of every bowler take remaing for the training and test sets.

first_inn_final_data=final_1st_inn_data[~final_1st_inn_data['total_balls_bowled'].isnull()]
```

```python
first_inn_final_data=first_inn_final_data.drop(['season','seasonal_match_id','overs_played','total_wickets_taken'],axis=1)
```

```python
first_inn_final_data['dismissal_kind']=(first_inn_final_data['dismissal_kind']).apply(int)
first_inn_final_data['total_balls_bowled']=(first_inn_final_data['total_balls_bowled']).apply(int)
first_inn_final_data['total_matches_played']=(first_inn_final_data['total_matches_played']).apply(int)
first_inn_final_data['total_runs_given']=(first_inn_final_data['total_runs_given']).apply(int)
```

```python
##### lable encoding of the teams.

teams_codes={'MI':0,
        'DD':1,
        'KKR':2,
        'SRH':3,
        'CSK':4,
        'RR':5,
        'RCB':6,
        'KXI':7,
        'DC':8,
        'KTK':9,
        'RPS':10,
        'GL':11,
        'PW':12,
}

first_inn_final_data=first_inn_final_data.replace({'batting_team':teams_codes,'bowling_team':teams_codes})
```

```python
#### Unique id create using match_id,bowler_id and bowling_team.

first_inn_final_data['unique_id']=(first_inn_final_data['match_id']).apply(str)+"_"+(first_inn_final_data['bowler_id']).apply(str)+"_"+(first_inn_final_data['bowling_team']).apply(str)
```

```python
### remove bowling_team, match_id, bowler_id.


del first_inn_final_data['match_id']
del first_inn_final_data['bowler_id']
del first_inn_final_data['bowling_team']
```

```python
##### seperate X and Y.


trainX=first_inn_final_data[['venue_id','batting_team','total_balls_bowled','total_matches_played','total_runs_given','bowling_average','bowling_strike_rate','cum_sum_overs']]
trainy=first_inn_final_data['dismissal_kind']
```

```python
#### creating dummy variables for venue_id and batting_team.


trainX['batting_team']='batting_team_'+(trainX['batting_team']).apply(str)
trainX['venue_id']='venue_id_'+(trainX['venue_id']).apply(str)
```

```python
dummy_venues=pa.get_dummies(trainX['venue_id'])
dummy_batting_team=pa.get_dummies(trainX['batting_team'])
```

```python
##### final data for modelling

X=pa.concat([dummy_venues.iloc[:,:-1],dummy_batting_team.iloc[:,:-1],trainX.iloc[:,2:]],axis=1)
y=trainy
```

```python
#### We can see this variables are skewed.So lets transform.

sns.distplot(X['total_balls_bowled'])
sns.distplot(X['total_runs_given'])
sns.distplot(X['cum_sum_overs'])
plt.show()
```

```python
#transformation of total_balls_bowled,total_runs_given

sns.distplot(np.sqrt(X['total_balls_bowled']))
sns.distplot(np.sqrt(X['total_runs_given']))

plt.show()
```

```python
sns.distplot(np.sqrt(X['cum_sum_overs']))
plt.show()
```

```python
X['cum_sum_overs']=np.sqrt(X['cum_sum_overs'])
X['total_balls_bowled']=np.sqrt(X['total_balls_bowled'])
X['total_runs_given']=np.sqrt(X['total_runs_given'])
```

```python
### Binning the target variable so that we can use it as classification problem.
### Whether the bolwer takes single wicket or 2 to 3 or more than 3.
### Four class classification 0,1,2,3,more than 3

### 0---0
### 1 and 2 --1
### more than 3  --2


y=np.array(y)


for i in range(len(y)):
    if y[i] ==0:
        y[i]=0
    elif y[i]>=1 and y[i]<=2:
        y[i]=1
    elif y[i]>=3:
        y[i]=2
```

```python
#### Random split of train and test set.

from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y = train_test_split(X,y,random_state=42,test_size=0.35)
```

```python
#### Classification algorithm.

from sklearn.ensemble import RandomForestClassifier


clf=RandomForestClassifier(n_estimators=1000,
                       criterion='gini',
                       max_features=None,
                       max_depth=3,min_samples_split=2,
                       min_samples_leaf=2,
                      )
```

```python
clf.fit(train_X,train_y)
```

```python
clf.predict(test_X)
```
