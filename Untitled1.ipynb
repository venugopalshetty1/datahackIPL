```python
match_wins = match_data['winner']

match_wins=pa.DataFrame(match_wins.value_counts())
match_wins=match_wins.reset_index()
match_wins.columns=['teams','wins']
```

```python
match_wins
```

```python
total_matchs_played=pa.Series(sum([list(match_data['team1_id']),list(match_data['team2_id'])],[]))

total_matchs_played=(pa.DataFrame(total_matchs_played.value_counts())).reset_index()
total_matchs_played.columns=['teams','matchs']
total_matchs_played=total_matchs_played.merge(match_wins, on="teams")
total_matchs_played['wining_percent']=(np.round((total_matchs_played['wins']/total_matchs_played['matchs'])*100,2)).apply(str)+"% Winning"
```

```python
total_matchs_played
```

```python
### This is the grouped bar chart based on number of matches played and number of matchs won over since 2008

#### Trace 0 Bars ---> number of matches played
trace0 = Bar(
x=total_matchs_played['teams'], y=total_matchs_played['matchs'], name="Total matches Played", marker=dict(color="rgba(55, 83, 109,0.5)",),
)

#### Trace 1 Bars ---> number of matches Won
trace1 = Bar(
x=total_matchs_played['teams'],y=total_matchs_played['wins'], name ="Total matches Won", marker=dict(color="rgba(26, 118, 255,0.5)"),
text=total_matchs_played['wining_percent']   
)


data= [trace0,trace1]
#### Designing layout means axis,legends styles are done in this object  


layout = Layout(
title="Number of Matches played VS Matches Won",
xaxis = dict(
    tickfont = dict(
        size=14,
        color="rgb(107, 107, 107)"
    )),
yaxis =dict(
    title="Number of matches",
    tickfont = dict(
        size = 15,
        color = "rgb(107,107,107)"
    )
),

)
plotly.offline.iplot({
    "data":data,
    "layout":layout
})
```

```python
batsman_details=ball_by_ball[['match_id','inning','batting_team','batsman','batsman_runs']]
batsman_details=batsman_details.merge(match_data[['match_id','season']],on="match_id")
```

```python
batsman_details.head()
```

```python
batsman_runs=batsman_details.groupby(['match_id','season','inning','batting_team','batsman'])['batsman_runs'].agg({'balls_played':'count','batsman_runs':'sum'}).reset_index()
```

```python
#### Calculate the strike rate for every season than every match


batsman_runs_season=batsman_runs.groupby(['batsman'])['balls_played','batsman_runs'].agg({'balls_played':'sum','batsman_runs':'sum'}).reset_index()
```

```python
batsman_runs_season['strick_rate']=np.round(batsman_runs_season['batsman_runs']/batsman_runs_season['balls_played']*100,2)
```

```python
from plotly.widgets import GraphWidget
from ipywidgets import widgets
from plotly.graph_objs import Scatter
```

```python
data=[Scatter(x=batsman_runs_season['batsman'],y=batsman_runs_season['strick_rate']
             ,mode="markers", marker=dict(color='rgb(0, 255, 193)', size=10), line=dict(width=8, color='rgb(255,0,0)')),
     ]
```

```python
plotly.offline.iplot({
    "data":data
})
```

```python
def plot_scatter(year):
    plotly.offline.iplot({
        "data":[Scatter(x=batsman_runs_season[batsman_runs_season['season']==year]['batsman'],y=batsman_runs_season[batsman_runs_season['season']==year]['strick_rate']
                 ,mode="markers", marker=dict(color='rgb(0, 255, 193)', size=10), line=dict(width=8, color='rgb(255,0,0)')),
         ]
    })    
```

```python
w=widgets.Dropdown(options=list(batsman_runs_season['season'].unique()))
display(w)
w.observe(plot_scatter,names="selected_label")
```
